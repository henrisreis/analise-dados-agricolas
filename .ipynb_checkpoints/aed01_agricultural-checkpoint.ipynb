# Análise Exploratória de Dados

Faremos a análise exploratória de um conjunto de dados referente ao **preço de matérias-primas agrícolas entre os anos de 1990 e 2020**.

O conjunto de dados entitulado **Agricultural Raw Material Prices (1990-2020)** encontra-se disponível <a href="https://www.kaggle.com/datasets/kianwee/agricultural-raw-material-prices-19902020" target="_blank">aqui</a>.

Carregando o conjunto de dados em um dataframe utilizando Pandas:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('agricultural_raw_material.csv')
```

Para uma exploração inicial do conjunto de dados, utilizaremos a função `head()`:

```python
df.head()
```

```python
df.info()
```

Para verificar se o conjunto de dados possui registros nulos:

```python
df.isnull().sum()
```

Observamos que há vários atributos com valores vazios de registros.

Substituindo os caracteres '%', ',' e '-' dos registros por valores vazios:

```python
df.replace({'%': '', ',': '', '-': ''}, inplace=True, regex=True)
```

Substituindo os valores vazios dos registros por objetos do tipo NaN:

```python
df.replace('', np.nan, inplace=True)
```

Apagando os registros com valores NaN:

```python
df.dropna(inplace=True)
```

Verificando se os valores NaN foram resolvidos:

```python
df.isnull().sum()
```

Convertendo o tipo dos dados numéricos para float:

```python
lista = [coluna for coluna in df.columns.tolist() if coluna != 'Month']

df[lista] = df[lista].astype('float')
```

```python
df.info()
```

```python
df.head()
```

Alterando o tipo dos valores da coluna `Month` para `datetime`:

```python
df['Month'] = pd.to_datetime(df['Month'], format='%b%y')
```

```python
df.head()
```

Definindo a coluna `Month` como  índice para o conjunto de dados:

```python
df.set_index('Month', inplace=True)
```

```python
df.head()
```

Para visualização dos dados utilizaremos as bibliotecas matplotlib.pyplot e seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Vamos entender qual é a correlação entre os preços das matérias-primas.

## O que é correlação?

Fonte - https://medium.com/omixdata/estat%C3%ADstica-an%C3%A1lise-de-correla%C3%A7%C3%A3o-usando-python-e-r-d68611511b5a

Quando falamos em análise de dados, muitas vezes precisamos entender qual é a associação entre duas ou mais variáveis. Nesse caso, a análise de correlação é uma forma descritiva que mede se há e qual o grau de dependência entre variáveis, ou seja, o quanto uma variável interfere em outra, lembrando que essa relação de dependência pode ou não ser causal. Essa medida de grau de relação é medida através de coeficientes. Vamos focar no coeficiente de Pearson.

O coeficiente de Pearson, também chamado de “coeficiente de correlação produto-momento” ou chamado de “ρ de Pearson”, mede o grau de correlação através do cálculo de direção positiva ou negativa. Este coeficiente, normalmente representado por ρ assume apenas valores entre -1 e 1.

A análise de correlação vai retornar três possíveis cenários: (1) correlação positiva; (2) correlação negativa; e (3) não há correlação.

- **Correlação positiva**: quando duas variáveis que possuem correlação crescem ou decrescem juntas, ou seja, que possuem uma relação direta;
- **Correlação negativa**: quando duas variáveis que possuem correlação mas quando uma variável cresce a outra decresce, ou vice-versa;
- **Não ter correlação**: quando o crescimento ou decrescimento de uma variável não tem efeito sobre outra variável.

Como podemos interpretar os valores que ρ pode assumir?

- **ρ = 0,9 a 1 (positivo ou negativo)**: correlação muito forte;
- **ρ = 0,7 a 0,9 (positivo ou negativo)**: correlação forte;
- **ρ = 0,5 a 0,7 (positivo ou negativo)**: correlação moderada;
- **ρ = 0,3 a 0,5 (positivo ou negativo)**: correlação fraca;
- **ρ = 0 a 0,3 (positivo ou negativo)**: não possui correlação.

Agora vamos avaliar a correlação entre os preços das matérias-primas.

```python
materias_primas = [coluna for coluna in df.columns.tolist() if coluna != 'Month' and '%' not in coluna]
materias_primas
```

Para obtermos a matriz de correlação entre as variáveis:

```python
matriz_correlacao = df[materias_primas].corr()
matriz_correlacao
```

Utilizaremos o mapa de calor para uma avaliação visual da correlação entre as variáveis:

```python
plt.figure(figsize=(12,9))

sns.heatmap(data=matriz_correlacao, annot=True, fmt='.2f')

plt.show()
```

Verifica-se uma **correlação positiva forte (ρ = 0,89)** entre os preços da **Lã Grossa (Coarse Wool)** e da **Lã Fina (Fine Wool)**. Quanto mais próximo de 1 for o valor do coeficiente de Pearson explícito no mapa de calor, mais forte é a correlação entre os preços das matérias-primas. É importante lembrar que a correlação não significa que um evento causa outro. Mas isso indica que eles tem um padrão ou um relacionamento entre si.

A seguir, iremos analisar a correlação entre as variações percentuais dos preços das matérias-primas.

```python
variacao = [coluna for coluna in df.columns.tolist() if coluna != 'Month' and '%' in coluna]
variacao
```

```python
matriz_correlacao = df[variacao].corr()
```

```python
plt.figure(figsize=(12,9))

sns.heatmap(data=matriz_correlacao, annot=True, fmt='.2f')

plt.show()
```

Verifica-se que não há correlação entre as variações percentuais de preço das matérias primas, já que o coeficiente de pearson mantém-se menor que 0,3 em toda a matriz de correlação.

Plotando as séries temporais do preço e da variação percentual do preço de cada uma das matérias-primas:

```python
lista_preco_variacao = [coluna for coluna in df.columns.tolist() if coluna != 'Month']

lista_preco_variacao
```

```python
def separar_pares(lista):
    elemento_2 = lista.pop()
    elemento_1 = lista.pop()
    return [elemento_1, elemento_2]

lista_pares = []

while len(lista_preco_variacao) > 0:
    lista_pares.append(separar_pares(lista_preco_variacao))

lista_pares
```

```python
for i in range(len(lista_pares)):
    df[lista[i]].plot(figsize=(11,9), subplots=True, title=lista[i][0])
    plt.show()
```
